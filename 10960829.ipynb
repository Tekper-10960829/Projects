```python
import pandas as pd 
import matplotlib.pyplot as plt 
import matplotlib as mpl
import numpy as np
import seaborn as sns
```

```python
data_1 = pd.read_csv("data.csv")
data_1
```

```python
data_2 = data_1.fillna(data_1.mean(), inplace=False)
data_2
```

```python
data_2.duplicated()
```

```python
mode_data_2 = data_2.iloc[:, 1:].mode().iloc[0]
median_data_2 = data_2.iloc[:, 1:].median()
print("Mode:")
print(mode_data_2)
print("\nMedian:")
print(median_data_2)
```

```python
histogram_1 = sns.histplot(data=data_2, x='fixed acidity')

histogram_1.set_title("Histogram for fixed acidity")
```

```python
mean_1 = data_2['fixed acidity'].mean()


above_mean_count = data_2[data_2['fixed acidity'] > mean_1]['fixed acidity'].count()
below_mean_count = data_2[data_2['fixed acidity'] < mean_1]['fixed acidity'].count()


labels = ['Greater than Mean', 'Less than Mean']
sizes = [above_mean_count, below_mean_count]
colors = ['blue', 'grey']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Fixed Acidity Greater than and Less than Mean')
plt.show()
```
